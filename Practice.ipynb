```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15, 9)
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python
weight = 0.0
x = 1.2
goal_prediction = 7.31
num_iters = 30
alpha = 0.1
errors = []
deltas = []
derivs = []
```

```python
for i in range(num_iters):
    # 1. Calculate a prediction
    prediction = weight * x
    
    # 2. Calculate the error
    squared_error = (prediction - goal_prediction) ** 2
    
    # 3. Calculate the derivative
    delta = (prediction - goal_prediction)
    deriv = delta * x
    
    # 4. Update the weight
    weight -= alpha * deriv

    # This for error and better understanding visualization
    errors.append(squared_error)
    deltas.append(delta)
    derivs.append(deriv)
    
    print(f'Squared Error: {squared_error:.2f} -- Raw Error: {delta:.2f}')
    print(f'Delta: {delta:.2f} -- Deriv: {deriv:.2f}')
    print(f'Prediction: {prediction:.2f}')
    print('--------------------------------')
    
```

```python
fig, ax = plt.subplots(1,3)
ax[0].plot(range(num_iters), errors)
ax[1].plot(range(num_iters), deltas)
ax[2].plot(range(num_iters), derivs)
plt.show()
```
